```python
# 1 - Exibir as primeiras 5 linhas do DataFrame
import pandas as pd

df = pd.read_csv('criminalidade_brasil_exercicio.csv')
display(df.head(5))
```

```python
# 2 - Filtrar dados para exibir apenas crimes de um estado específico (exemplo: SP)
df_filtrado = df[df['Estado'] == 'MG']
display(df_filtrado.head())
```

```python
# 3 - Contar o número de ocorrências por tipo de crime e criar um gráfico de barras
import matplotlib.pyplot as plt

tipo_crimes_counts = df['Tipo_Crime'].value_counts()
tipo_crimes_counts.plot(kind='bar', color='green')
plt.xlabel('Tipo de Crime')
plt.ylabel('Quantidade')
plt.title('Número de Ocorrências por Tipo de Crime')
plt.show()
```

```python
# 4 - Criar um gráfico de linha mostrando a evolução dos crimes ao longo dos anos
crimes_por_ano = df.groupby('Ano')['Quantidade'].sum()
crimes_por_ano.plot(kind='line', marker='o', color='red')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Crimes')
plt.title('Evolução dos Crimes ao Longo dos Anos')
plt.show()
```
